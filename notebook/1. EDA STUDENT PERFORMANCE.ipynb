### 1) Problem statement
- This project understands how the student's performance (test scores) is affected by other variables such as Gender, Ethnicity, Parental level of education, Lunch and Test preparation course.


### 2) Data Collection
- Dataset Source - https://www.kaggle.com/datasets/spscientist/students-performance-in-exams?datasetId=74977
- The data consists of 8 column and 1000 rows.

### 3.8 Adding columns for "Total Score" and "Average"

```python
df["total_score"] = df["math_score"] + df["reading_score"] + df["writing_score"]
df["average_score"] = df["total_score"] / 3
df.head()
```

```python
reading_full = df[df['reading_score'] == 100]['average_score'].count()
writing_full = df[df['writing_score'] == 100]['average_score'].count()
math_full = df[df['math_score'] == 100]['average_score'].count()

print("Number of students with full marks in reading: ", reading_full)
print("Number of students with full marks in writing: ", writing_full)
print("Number of students with full marks in math: ", math_full)
```

```python
reading_less_20 = df[df['reading_score'] <= 20]['average_score'].count()
writing_less_20 = df[df['writing_score'] <= 20]['average_score'].count()
math_less_20 = df[df['math_score'] <= 20]['average_score'].count()

print(f'Number of students with less than 20 marks in Maths: {math_less_20}')
print(f'Number of students with less than 20 marks in Writing: {writing_less_20}')
print(f'Number of students with less than 20 marks in Reading: {reading_less_20}')
```

#####  Insights
 - From above values we get students have performed the worst in Maths 
 - Best performance is in reading section

#### 4.4.2 RACE/EHNICITY COLUMN
- How is Group wise distribution ?

#### UNIVARIATE ANALYSIS ( How is Group wise distribution ?)

```python
fig, ax = plt.subplots(1, 2, figsize=(16, 6))

colors = ['#ff9999', '#66b3ff', '#99ff99', '#ffcc99', '#c2c2f0']

sns.countplot(data=df, x='race_ethnicity', palette=colors, saturation=0.95,
    order=df['race_ethnicity'].value_counts().index, ax=ax[0])
for container in ax[0].containers:
    ax[0].bar_label(container)

ax[0].set_title('Race/Ethnicity Count')
ax[0].set_xlabel('Race/Ethnicity')
ax[0].set_ylabel('Count')
ax[0].tick_params(axis='x', rotation=30)

# Pie chart
counts = df['race_ethnicity'].value_counts()
ax[1].pie(counts, labels=counts.index, colors=colors, explode=[0.1, 0, 0, 0, 0],
    autopct='%1.1f%%', shadow=True, startangle=90)
ax[1].set_title('Race/Ethnicity Distribution')
ax[1].axis('equal')

plt.tight_layout()
plt.show()
```

#### BIVARIATE ANALYSIS 
( Is Race/Ehnicity has any impact on student's performance ? )

```python
group_data = df.groupby('race_ethnicity')[['math_score', 'reading_score', 'writing_score']].mean()

fig, ax = plt.subplots(1, 3, figsize=(20, 8))

# Math score
sns.barplot(x=group_data.index, y=group_data['math_score'], palette='mako',ax=ax[0])
ax[0].set_title('Average Math Score', color='#005ce6', fontsize=18)
ax[0].set_xlabel('Race/Ethnicity')
ax[0].set_ylabel('Score')

for container in ax[0].containers:
    ax[0].bar_label(container, fmt='%.1f')

# Reading score
sns.barplot(x=group_data.index,y=group_data['reading_score'],palette='flare',ax=ax[1])
ax[1].set_title('Average Reading Score', color='#005ce6', fontsize=18)
ax[1].set_xlabel('Race/Ethnicity')
ax[1].set_ylabel('Score')
for container in ax[1].containers:
    ax[1].bar_label(container, fmt='%.1f')

# Writing score
sns.barplot(x=group_data.index,y=group_data['writing_score'], palette='coolwarm',ax=ax[2])
ax[2].set_title('Average Writing Score', color='#005ce6', fontsize=18)
ax[2].set_xlabel('Race/Ethnicity')
ax[2].set_ylabel('Score')
for container in ax[2].containers:
    ax[2].bar_label(container, fmt='%.1f')

# Rotate x-axis labels
for a in ax:
    a.tick_params(axis='x', rotation=30)

plt.tight_layout()
plt.show()
```

#### Insights 
- Group E students have scored the highest marks. 
- Group A students have scored the lowest marks. 
- Students from a lower Socioeconomic status have a lower avg in all course subjects
